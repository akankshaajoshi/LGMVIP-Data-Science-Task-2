# **Image to Pencil Sketch**

Create the pencil sketch of an image in the given directory using OpenCV and Python.

```python
#Importing the libraries

import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
```

```python
#Entering the location of image input

img_loc = input("Enter the complete location of the image file:")
```

```python
#Read the image using the imread option

img = cv2.imread(img_loc)
```

```python
#Show the image

cv2_imshow(img)
```

```python
#Convert the color from BGR to GRAY

grey_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
#Check the converted image

cv2_imshow(grey_img)
```

```python
#Invert the image

inverted = 255 - grey_img
```

```python
#Check the inverted image

cv2_imshow(inverted)
```

```python
#Blur the inverted image using the highest filter size

blur_img=cv2.GaussianBlur(inverted, (111,111),0)
```

```python
#Invert the blurred image

inverted_blur = 255 - blur_img
```

```python
#Show the inverted blurred image

cv2_imshow(inverted_blur)
```

```python
#Generate a sketch by dividing the grayscale image to the inverted blur image

sketch = cv2.divide(grey_img, inverted_blur, scale = 256.0)
```

```python
#Show the final result

cv2_imshow(sketch)
```

```python
#Save the result

cv2.imwrite('sketch.jpg', sketch)
```
